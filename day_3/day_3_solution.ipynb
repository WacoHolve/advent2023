```python
from pandas import read_table

df = read_table("day_3_test.txt")
```

```python
def _is_period_or_numeric(val: str or int) -> bool:
    if not val:
        return True

    if (val.isnumeric()) or (val == "."):
        return True
    return False

def _search_left(entries: str, column: int) -> str:
    value = str(entries[column])
    numeric_value = True
    while numeric_value:
        try: 
            new_value = entries[column - 1]
            if new_value.isnumeric():
                value = str(new_value) + value
            if not new_value.isnumeric():
                numeric_value = False
        except:
            numeric_value = False
        
        column -= 1
    return value

def _search_right(entries: str, column: int) -> str:
    value = str(entries[column])
    numeric_value = True
    while numeric_value:
        try: 
            new_value = entries[column + 1]
            if new_value.isnumeric():
                value += str(new_value)
            if not new_value.isnumeric():
                numeric_value = False
        except:
            numeric_value = False
        
        column += 1
    return value

all_rows = len(df) - 1
solution = 0
for row in range(len(df)):
    entries = df.iat[row, 0]
    valid_parts = []
    for column in range(len(entries)):
        val = entries[column]
        row_len = len(entries) - 1
        if val.isnumeric():
            above_row = row - 1
            below_row = row + 1
            left_column = column - 1
            right_column = column + 1
            above = None
            top_left = None
            top_right = None
            bottom = None
            bottom_left = None
            bottom_right = None
            left = None
            right = None
            if row > 0:
                above = df.iat[above_row, 0][column]
                if column > 0:
                    top_left = df.iat[above_row, 0][left_column]
                if column < row_len:
                    top_right = df.iat[above_row, 0][right_column]
            if row < all_rows:
                bottom = df.iat[below_row, 0][column]
                if column > 0:
                    bottom_left = df.iat[below_row, 0][left_column]
                if column < row_len:
                    bottom_right = df.iat[below_row, 0][right_column]
            if column > 0:
                left = df.iat[row, 0][left_column]
            if column < row_len:
                right = df.iat[row, 0][right_column]
        

            valid_part_number = False
            for cell in [above, top_left, top_right, bottom, bottom_left, bottom_right, left, right]:
                if not _is_period_or_numeric(val=cell):
                    valid_part_number = True
                    left_value = _search_left(entries=entries, column=column)
                    right_value = _search_right(entries=entries, column=column)

                    part_number = left_value + right_value[1:]
                    if part_number.isnumeric():
                        valid_parts.append(int(part_number))
                        last_number = column

    print((valid_parts))
    solution += sum((valid_parts))
            
solution
```
