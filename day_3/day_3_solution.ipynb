```python
from pandas import read_table

def _is_period_or_numeric(val: str or int) -> bool:
    if not val:
        return True

    if (val.isnumeric()) or (val == "."):
        return True
    return False

def _search_left(entries: str, column: int) -> str:
    value = str(entries[column])
    numeric_value = True
    while numeric_value:
        try: 
            new_value = entries[column - 1]
            if new_value.isnumeric():
                value = str(new_value) + value
            if not new_value.isnumeric():
                numeric_value = False
        except:
            numeric_value = False
        
        column -= 1
    return value

def _search_right(entries: str, column: int) -> str:
    value = str(entries[column])
    numeric_value = True
    while numeric_value:
        try: 
            new_value = entries[column + 1]
            if new_value.isnumeric():
                value += str(new_value)
            if not new_value.isnumeric():
                numeric_value = False
        except:
            numeric_value = False
        
        column += 1
    return value

def _extract_number(entries: str, column: int) -> int:
    left = _search_left(entries=entries, column=column)
    right = _search_right(entries=entries, column=column)

    return int(left + right[1:])

def _is_special_character_adjacent(
        row: int, column: int, entries: str, all_rows: int
) -> bool:
    above_row = row - 1
    below_row = row + 1
    left_column = column - 1
    right_column = column + 1
    row_len = len(entries) - 1

    above = None
    top_left = None
    top_right = None
    bottom = None
    bottom_left = None
    bottom_right = None
    left = None
    right = None

    if row > 0:
        above = df.iat[above_row, 0][column]
        if column > 0:
            top_left = df.iat[above_row, 0][left_column]
        if column < row_len:
            top_right = df.iat[above_row, 0][right_column]
    if row < all_rows:
        bottom = df.iat[below_row, 0][column]
        if column > 0:
            bottom_left = df.iat[below_row, 0][left_column]
        if column < row_len:
            bottom_right = df.iat[below_row, 0][right_column]
    if column > 0:
        left = df.iat[row, 0][left_column]
    if column < row_len:
        right = df.iat[row, 0][right_column]
    
    for cell in [above, top_left, top_right, bottom, bottom_left, bottom_right, left, right]:
        if not _is_period_or_numeric(val=cell):
            return True
    return False

df = read_table("day_3_input.txt")
solution = 0
all_rows = df.shape[0] - 1
for row in range(len(df)):
    entry = df.iat[row, 0]

    valid_parts = []
    ran_into_period = True
    for column in range(len(entry)):
        val = entry[column]
        if val == "." or not val.isnumeric():
            ran_into_period = True

        if val.isnumeric() and _is_special_character_adjacent(
            row=row, column=column, entries=entry, all_rows=all_rows
        ) and ran_into_period:
            ran_into_period = False
            part_number = _extract_number(entries=entry, column=column)
            solution += part_number

solution
```
